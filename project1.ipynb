```python
#Removing Duplicates
df_wal.drop_duplicates(inplace= True)

#Validating the renoval of Duplicates
df_wal.duplicated().sum()
```

```python
#Removing the null values. Rather than filling the values I'm preferring to remove the values

df_wal.dropna(inplace=True)

#rechecking whether the null values hae been removed/dropped
df_wal.isnull().sum()
```

```python
df_wal.info()
```

```python
#Converting the the unit price from text to float
df_wal["unit_price"].astype(float)
```

```python
df_wal["unit_price"]
```

```python
df_wal["unit_price"] = df_wal["unit_price"].str.replace('$','')
df_wal["unit_price"] = df_wal["unit_price"].astype(float)
```

```python
df_wal["category"].nunique() #gives 6 unique values
df_wal["Branch"].nunique() #gives us 98 unique values
df_wal["Branch"].nunique() #gives 100 unique values
df_wal["payment_method"].nunique() #gives 3 unique values

# We can therefor convert these less unique values from string type to category type as this helps us better optimization of memory
# The category type is ideal for columns with a limited number of unique values.
df_wal["category"] = df_wal["category"].astype("category")
df_wal["payment_method"] = df_wal["payment_method"].astype("category")

#Now re checking df_info method you will see better memory usage with these 2 columns dtypes being updated to category
df_wal.info()

#the memory usage previously was: 863.9+ KB , now the memory usage: 798.6+ KB
```

```python
#Adding a new column as total price which is simple unit_price * quantity 

df_wal["total_price"] = df_wal["unit_price"] * df_wal["quantity"]
```

```python
#now checking the dataset after removing the nulls and duplicates and adding a new column
df_wal.head()
```
